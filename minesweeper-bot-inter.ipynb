```python
import pyautogui
import time
import cv2
import numpy as np

# Função para clicar com o botão esquerdo do mouse
def click_left(x, y):
    pyautogui.click(x, y)

# Função para verificar se a cor é igual a um dos alvos
def is_color_equal(color, target_color):
    return color == target_color

# Função para reiniciar o jogo
def reinicia():
    color = pyautogui.pixel(782, 599)
    if is_color_equal(color, (113, 63, 18)):
        click_left(782, 599)
        time.sleep(0.01)
        click_left(745, 217)
        time.sleep(0.01)
        click_left(759, 375)
        return True
    else:
        return False

# Cores
ignore_1 = (55, 65, 81)
ignore_2 = (75, 85, 99)
flag_1 = (2, 6, 23)
flag_2 = (30, 41, 59)
target_1 = (87, 83, 78)
target_2 = (68, 64, 60)

# Distâncias
square_distance_x = 28
square_distance_y = 28

# Tamanho do tabuleiro
board_size_x = 9
board_size_y = 9

# Posição inicial do mouse
start_x = 638
start_y = 263

# Contador de iterações
iteration_counter = 0

# Última posição clicada
last_clicked_position = None

time.sleep(2)

while True:
    # Inicializa o contador para a iteração atual
    current_iteration_counter = 0

    for y in range(board_size_y):
        for x in range(board_size_x):
            # Calcula as coordenadas do centro do quadrado
            square_center_x = start_x + x * square_distance_x
            square_center_y = start_y + y * square_distance_y

            # Obtém a cor do quadrado no centro diretamente da imagem capturada
            color = pyautogui.pixel(square_center_x, square_center_y)

            # Verifica se a cor do quadrado é igual a um dos alvos
            if is_color_equal(color, target_1) or is_color_equal(color, target_2):
                click_left(square_center_x, square_center_y)
                current_iteration_counter += 1

                # Atualiza a última posição clicada
                last_clicked_position = (square_center_x, square_center_y)

    if reinicia():
        last_clicked_position = None
        iteration_counter = 0  # Zera o contador ao reiniciar o jogo
        print("Jogo reiniciado!")

    # Verifica se o contador da iteração atual é maior que 1
    if current_iteration_counter > 1:
        # Incrementa o contador global
        iteration_counter += 1

        # Verifica se a última posição clicada se repetiu 5 vezes
        if last_clicked_position is not None and iteration_counter == 16:
            last_clicked_position = None
            iteration_counter = 0  # Zera o contador ao reiniciar o jogo
            click_left(745, 217)
            time.sleep(0.3)
            click_left(945, 586)
            time.sleep(0.3)
            click_left(759, 375)

        # Verifica se a última posição clicada se repetiu 8 vezes e clica no quadrado ao redor
        elif last_clicked_position is not None and (iteration_counter == 9 or iteration_counter == 11):
            click_left(last_clicked_position[0] - square_distance_x, last_clicked_position[1] + square_distance_y)
            click_left(last_clicked_position[0] + square_distance_x, last_clicked_position[1] - square_distance_y)
            last_clicked_position = None
        elif last_clicked_position is not None and (iteration_counter == 8 or iteration_counter == 10):
            click_left(last_clicked_position[0] - square_distance_x, last_clicked_position[1] - square_distance_y)
            click_left(last_clicked_position[0] + square_distance_x, last_clicked_position[1] + square_distance_y)
            last_clicked_position = None
    else:
        if reinicia():
            last_clicked_position = None
            iteration_counter = 0  # Zera o contador ao reiniciar o jogo
            print("Jogo reiniciado!")
        else:
            last_clicked_position = None
            iteration_counter = 0  # Zera o contador ao reiniciar o jogo
            click_left(745, 217)
            time.sleep(0.3)
            click_left(945, 586)
            time.sleep(0.3)
            click_left(759, 375)
    print(f"Iteração {iteration_counter}")
```
