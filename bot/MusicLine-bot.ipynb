```python
import pyautogui
import time
import keyboard
import win32api
import win32con
import concurrent.futures

time.sleep(2)

def click():
    win32api.SetCursorPos((688, 789))
    win32api.mouse_event(win32con.MOUSEEVENTF_LEFTDOWN, 0, 0)
    win32api.mouse_event(win32con.MOUSEEVENTF_LEFTUP, 0, 0)
    print("click")

def check_and_click(x, y, pic):
    r, g, b = pic.getpixel((x, y))
    if r == 245 and g == 17 and b == 9:
        click()
        return True
    return False

while not keyboard.is_pressed('q'):
    flag = 0
    pic = pyautogui.screenshot(region=(7, 620, 944, 405))
    pic.save(r".\game.png")
    width, height = pic.size

    with concurrent.futures.ThreadPoolExecutor() as executor:
        futures = []

        # Largura: 602
        for x in range(0, width, 2):
            # Altura: 418
            for y in range(0, height, 2):
                futures.append(executor.submit(check_and_click, x, y, pic))

        # Get the results as tasks complete
        for future in concurrent.futures.as_completed(futures):
            if future.result():
                flag = 1
                break

    if flag == 1:
        time.sleep(0.1)
    else:
        break
```
